```python
import json
from datetime import datetime

file = "linkedin_data/posts_20260228_140507.json"

with open(file, "r") as f:
    data = json.load(f)

datetime.fromtimestamp(min([x["estimated_upload_time"] for x in data]))

min_date = None
idx = None
for i, x in enumerate(data):
    if min_date is None or x["estimated_upload_time"] < min_date:
        min_date = x["estimated_upload_time"]
        idx = i

print(f"Min date: {datetime.fromtimestamp(min_date)}")
print(f"Index: {idx}")
print(f"Post: {data[idx]}")
print(f"Count post: {len(data)}")
```
